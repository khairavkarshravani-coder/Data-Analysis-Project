### Data Loading:

### The dataset contains 418 entries and 12 columns.Each row represents an individual passenger.
    Survived: The target variable (0= No, 1= Yes).
     Pclass: Passenger Class (1 = Upper, 2= Middle, 3= Lower).
    Sex: Passenger's gender.
    Age: Age in years.
    SibSp / Parch: Indicates family members aboard (Siblings/Spouses/Parents/Children).
    Fare: The price paid for the ticket.
    Embarked: The port of boarding (C, Q, or S).
    C = Cherbourg, Q = Queenstown, S = Southampton)

# Insights:

    - Female passengers had a higher survival rate than male passengers.
    - First Class passengers were more likely to survive compared to passengers in other classes.
    - Age was an important factor, and children had a better chance of survival.
    -Passengers who paid higher fares had a greater probability of survival.
    -Handling missing values was very important for accurate data analysis.

```python
print("Step: Calculating Survival Rate by Gender...\n")
gender_survival = df.groupby("Sex")["Survived"].mean()
print("Survival Rate (Male vs Female):")
print(gender_survival)
print("\nStep: Plotting Bar Chart...\n")
gender_survival.plot(kind="bar")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

 ### Females had a higher survival probability

# Survival Rate by Passenger Class

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
print("Step: Calculating Survival Rate by Passenger Class...\n")
class_survival = df.groupby("Pclass")["Survived"].mean()
print("Survival Rate by Class (1st, 2nd, 3rd):")
print(class_survival)
print("\nStep: Plotting Bar Chart...\n")
class_survival.plot(kind="bar")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

 ### First class passengers survived more

# Age Distribution

```python
print("Step: Analyzing Age Distribution...\n")
print("Age Column Summary:")
print(df["Age"].describe())
print("\nStep: Plotting Age Distribution Histogram...\n")
df["Age"].hist()
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
print("Step: Calculating Correlation Matrix...\n")
corr_matrix = df.corr(numeric_only=True)
print("Correlation Matrix:")
print(corr_matrix)
print("\nStep: Plotting Correlation Heatmap...\n")
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

 # 5. Data Visualization:(Plotting Distribution Plots)


### Visualization helps us understand the data and identify patterns. We can create various types of plots using libraries like Matplotlib and Seaborn.

```python
###CountPlot
sns.catplot(data=df, x="Sex",hue="Survived",kind="count")
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Passenger Class')
plt.savefig('class_survival.png')
```

```python
# Divide Fare into 4 bins
df['Fare'] = pd.qcut(df['Fare'], 4)

# Barplot - Shows approximate values based 
# on the height of bars.
sns.barplot(x ='Fare', y ='Survived', data = df)
```

```python
# Countplot
sns.catplot(x ='Embarked', hue ='Survived', 
kind ='count', col ='Pclass', data = df)
```

```python
# Group the dataset by Pclass and Survived and then unstack them
group = df.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()

# Heatmap - Color encoded 2D representation of data.
sns.heatmap(pclass_survived, annot = True, fmt ="d")
```

```python
sns.barplot(data=df,x="Sex",y="Age")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title('Correlation Heatmap')
plt.show()
```
